# Creating a dash plot

```python
import dash             
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output

import plotly   
import plotly.express as px
```

```python
mark_values = {2012:'2012',2013:'2013',2014:'2014',2015:'2015',
               2016:'2016'
               }
```

```python
app = dash.Dash(__name__)
app.layout = html.Div([
##################################################################################################       
    html.Div([
            html.Pre(children= "Invoice count and Total revenue generation of States yearly",
            style={"text-align": "center", "font-size":"100%", "color":"black"})
        ]),

        html.Div([
            dcc.Graph(id='the_graph')
        ]),

        html.Div([
            dcc.RangeSlider(id='the_year',
                min=2012,
                max=2016,
                value=[2012,2013],
                marks=mark_values,
                step=None)
        ],style={"width": "90%", "position":"absolute",
                 "left":"5%"})
#################################################################################################
    
###################################################################################################

])

###calling back the funcions
@app.callback(
    Output('the_graph','figure'),
    [Input('the_year','value')])

    
def update_graph(years_chosen):
    #print(years_chosen)

    dff=graph_df[(graph_df['year']>=years_chosen[0])&(graph_df['year']<=years_chosen[1])]
    # filter df rows where column year values are >=1985 AND <=1988
   # dff=dff.groupby(["country"], as_index=False)["suicides/100k pop",
               #     "gdp_per_capita ($)"].mean()
    # print (dff[:3])

    scatterplot = px.scatter(
        data_frame=dff,
        x="invoice count",
        y="Total Amt Wtd Tax. sum",
        hover_data=['State','year'],
       # mode = "markers",
        #marker = dict(color = 'rgba(255, 128, 255, 0.8)'),
        text="State",
        template='plotly',
        height=550
    )

    scatterplot.update_traces(textposition='top center')

    return (scatterplot) 
###############################################################################################

##################################################################
```

```python
if __name__ == '__main__':
    app.run_server()
```

# Order type analysis of each state

```python
fakhru=invoice_data[['District','Order Type']]
```

```python
df0=pd.DataFrame(fakhru.groupby('District')['Order Type'].value_counts().sort_values(ascending=False))
df0=df0.rename(columns={'Order Type':'count'})
df0=df0.reset_index()
df0=df0.sort_values(by='District')
```

```python
df0.head()
```

```python
app = dash.Dash(__name__)
#############################################################################################
app.layout= html.Div([
    html.Div([
        html.Label(['Order type analysis of States']),
        dcc.Dropdown(
            id='my_dropdown',
            options=[{'label':i,'value':i} for i in df0.District.unique()],
            value='Andhra Pradesh',
            multi=False,
            clearable=False,
            style={'width':'60%'}
        ),
    ]),

    html.Div([
        dcc.Graph(id='the_graph',
                 style={'width':'60%'}
                 )
    ]),

])
###########################################################################################
@app.callback(Output('the_graph', 'figure'),
              [Input('my_dropdown', 'value')])

def update_figure(selected_state):
    filtered_df = df0[df0['District'] == selected_state]
    traces=[]
    #for selected_state in df1.State.unique():
    
    trace1 = go.Bar(
             x = filtered_df['Order Type'],
             y=filtered_df['count'],   
            name = selected_state,
            marker = dict(color = 'rgba(125, 125, 255, 0.5)',
            line=dict(color='rgb(0,0,0)',width=1.5)))
    traces.append(trace1)
    
    

    return {
        'data': traces,
        'layout': go.Layout(
            title='Order Type count State recieve',
            xaxis={ 'title': 'Order Type'},
            yaxis={'title': 'Count'},
            hovermode='closest')
        }
if __name__ == '__main__':
    app.run_server()
```
