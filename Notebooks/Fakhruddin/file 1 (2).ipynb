```python
cars=pd.DataFrame(df['Make'].value_counts(normalize=True)*100)
cars=cars.reset_index()
cars=cars.rename(columns={'index':'Car Brands','Make':'Percentage'})
```

```python
cars
```

```python
import plotly.express as px
# This dataframe has 244 lines, but 4 distinct values for `day`
#df = px.data.tips()
colors = ['gold', 'mediumturquoise', 'darkorange', 'lightgreen']
fig = px.pie(cars.head(10), values='Percentage', names='Car Brands',title='Top 10 Car brands across States:')
fig.update_traces(hoverinfo='label+percent', textinfo='value', textfont_size=10,
                 marker=dict(colors=colors, line=dict(color='#000000', width=1)))
fig.show()
```
