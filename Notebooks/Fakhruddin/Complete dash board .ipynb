# Analysis on the basis of make and model revenue generation

```python
selected_df=master_df[['Make','Model','Total_Expense']]
```

```python
selected_df.head()
```

```python
make_df=selected_df.groupby(['Make','Model'])['Total_Expense'].sum().sort_values(ascending=False).to_frame()
```

```python
make_df=make_df.reset_index()
make_df.head()
```

# Analysis of Order type how it is varying in different states:

```python
selected_ordertype=master_df[['State','Order Type']]
selected_ordertype.head()
```

```python
ordertype=pd.DataFrame(selected_ordertype.groupby('State')['Order Type'].value_counts().sort_values(ascending=False))
ordertype=ordertype.rename(columns={'Order Type':'count'})
ordertype=ordertype.reset_index()
ordertype=ordertype.sort_values(by='State')
```

```python
ordertype.head()
```

# Analysis of revenue and invoice in different states 

```python
invoice_data=pd.read_csv('C:\\Users\\Fakhruddin\\Capstone Project\\Mahindra First Choice Dataset\\Final_invoice.csv')
```

```python
invoice=invoice_data[['District','Invoice No','Invoice Date','Invoice Time','Total Amt Wtd Tax.']]
```

```python
invoice['invoice_datetime']=invoice['Invoice Date']+' '+invoice['Invoice Time']
invoice['invoice_datetime']=pd.to_datetime(invoice['invoice_datetime'])
```

```python
invoice=invoice.drop(['Invoice Date','Invoice Time'],axis=1)
```

```python
invoice['year']=invoice['invoice_datetime'].dt.year
invoice=invoice.rename(columns={'District':'State'})
```

```python
grp=invoice.groupby(['State','year'])
graph_df=grp.agg({'Invoice No':['count'],'Total Amt Wtd Tax.':'sum'}).sort_values(by='State')
graph_df
```

```python
graph_df=graph_df.reset_index()
graph_df.tail(10)
```

```python
graph_df.columns=['State','year','invoice count','Total Amt Wtd Tax. sum']
graph_df
```

```python
mark_values = {2012:'2012',2013:'2013',2014:'2014',2015:'2015',
               2016:'2016'
               }
```

# Analysis of  how car brands generate revenue  in each state 

```python
df2=master_df[['State','Make','Total_Expense']]
```

```python
df2=df2.groupby(['State','Make'])['Total_Expense'].sum().sort_values(ascending=False).to_frame()
```

```python
df2=df2.reset_index()
df2.head()
```

```python
import dash             
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output

import plotly   
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
```

```python
app = dash.Dash(__name__)
server=app.server

####################################################################################################
#Layout of the dashboard:
app.layout= html.Div([
    html.H1(
        children='EDA DashBoard',
        style={
            'textAlign': 'center',
            'color': '#00FF7F'
        }
    ),
########################################################################################################
#################layout for Make and Model revenue Analysis###################################   
    html.Div([
        html.Label(['Make and Model revenue Analysis']),
        dcc.Dropdown(
            id='my_dropdown',
            options=[{'label':i,'value':i} for i in make_df.Make.unique()],
            value='mahindra',
            multi=False,
            clearable=False,
            style={'width':'50%'}
        ),
    ]),

    html.Div([
        dcc.Graph(id='the_graph',
                 style={'width':'70%'}
                 )
    ]),
##############################layout for Based on Make State revenue Analysis######################################
      html.Div([
        html.Label(['Based on Make State revenue Analysis']),
        dcc.Dropdown(
            id='my_staterevenue',
            options=[{'label':i,'value':i} for i in df2.State.unique()],
            value='maharashtra',
            multi=False,
            clearable=False,
            style={'width':'50%'}
        ),
    ]),

    html.Div([
        dcc.Graph(id='State_Make_revenue_graph',
                 style={'width':'70%'}
                 )
    ]),
#############################################################################################################    
    
    

######################layout for Invoice count and Total revenue generation of States yearly #############################
      html.Div([
            html.Pre(children= "Invoice count and Total revenue generation of States yearly",
            style={"text-align": "center", "font-size":"100%", "color":"black"})
        ]),

        html.Div([
            dcc.Graph(id='scatter_graph')
        ]),

        html.Div([
            dcc.RangeSlider(id='the_year',
                min=2012,
                max=2016,
                value=[2012,2013],
                marks=mark_values,
                step=None)
        ],style={"width": "90%", "position":"absolute",
                 "left":"5%"}),
    html.Br(),
    html.Br(),
    html.Br(),
###################layout for Analysis of Order type in different state ######################################
     html.Div([
        html.Label(['Analysis of Order type in different state']),
        dcc.Dropdown(
            id='ordertype_dropdown',
            options=[{'label':i,'value':i} for i in ordertype.State.unique()],
            value='andhra pradesh',
            multi=False,
            clearable=False,
            style={'width':'50%'}
        )
    ]),

    html.Div([
        dcc.Graph(id='ordertype_graph',
                 style={'width':'70%'}
                 )
            ])
    
    
        ])

###################################################################################################

    
####################################################################################################    


###############################For Make and Model revenue generation##############################
#calling back the output and graphs
@app.callback(Output('the_graph', 'figure'),
              [Input('my_dropdown', 'value')])


##################################################################################################
#defining a function for make and model:
def update_figure(selected_make):
    filtered_df = make_df[make_df['Make'] == selected_make]
    traces=[]
    #for selected_state in df1.State.unique():
    
    trace1 = go.Bar(
             x = filtered_df['Model'],
             y=filtered_df['Total_Expense'],   
            name = selected_make,
            marker = dict(color = 'rgba(127,255,0, 0.5)',
            line=dict(color='rgb(0,0,0)',width=1.5)))
    traces.append(trace1)
    
    

    return {
        'data': traces,
        'layout': go.Layout(
            title='Make and model revenue analysis',
            xaxis={ 'title': 'Model'},
            yaxis={'title': 'Total revenue generated'},
            hovermode='closest')
        }
#########################callback and function for Based on Make State revenue Analysis ###################
@app.callback(Output('State_Make_revenue_graph', 'figure'),
              [Input('my_staterevenue', 'value')])
def update_figure(selected_state):
    filtered_df2 = df2[df2['State'] == selected_state]
    traces=[]
    #for selected_state in df1.State.unique():
    
    trace1 = go.Bar(
             x = filtered_df2['Make'],
             y=filtered_df2['Total_Expense'],   
            name = selected_state,
            marker = dict(color = 'rgba(239,222,205, 0.7)',
            line=dict(color='rgb(0,0,0)',width=1.5)))
    traces.append(trace1)
    
    

    return {
        'data': traces,
        'layout': go.Layout(
            title='Revenue Generated in a State on the basis of Car Brands ',
            xaxis={ 'title': 'Car Brands'},
            yaxis={'title': 'Revenue'},
            hovermode='closest')
        
    

        }

###########################################################################################################


#############################for invoice and revenue of state comparision over years##############################################
#callback for scatter plot
@app.callback(
    Output('scatter_graph','figure'),
    [Input('the_year','value')])
#defining function for scatter plot    
def update_graph(years_chosen):
    #print(years_chosen)

    dff=graph_df[(graph_df['year']>=years_chosen[0])&(graph_df['year']<=years_chosen[1])]
    # filter df rows where column year values are >=1985 AND <=1988
   # dff=dff.groupby(["country"], as_index=False)["suicides/100k pop",
               #     "gdp_per_capita ($)"].mean()
    # print (dff[:3])

    scatterplot = px.scatter(
        data_frame=dff,
        x="invoice count",
        y="Total Amt Wtd Tax. sum",
        hover_data=['State','year'],
       # mode = "markers",
        #marker = dict(color = 'rgba(255, 128, 255, 0.8)'),
        text="State",
        template='simple_white',
        height=550
    )

    scatterplot.update_traces(textposition='top center')

    return (scatterplot)

################################################################################################################
###########################For state wise order distribution#############################################
##callback for order type
@app.callback(
    Output('ordertype_graph','figure'),
    [Input('ordertype_dropdown','value')])
###defining function for selected state order type:
def update_figure(selected_state):
    filtered_df1 = ordertype[ordertype['State'] == selected_state]
    traces0=[]
    #for selected_state in df1.State.unique():
    
    trace2 = go.Bar(
             x = filtered_df1['Order Type'],
             y=filtered_df1['count'],   
            name = selected_state,
            marker = dict(color = 'rgba(125, 125, 255, 0.5)',
            line=dict(color='rgb(0,0,0)',width=1.5)))
    traces0.append(trace2)
    
    

    return {
        'data': traces0,
        'layout': go.Layout(
            title='Order Type count State recieve',
            xaxis={ 'title': 'Order Type'},
            yaxis={'title': 'Count'},
            hovermode='closest')
        }



if __name__ == '__main__':
    app.run_server()
```
